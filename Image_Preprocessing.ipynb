```python
# ---------------------------
# 1. LOAD IMAGE DATASET
# ---------------------------

import cv2
import numpy as np
import matplotlib.pyplot as plt

from google.colab import files
uploaded = files.upload()

# Load uploaded image (first image)
for filename in uploaded.keys():
    img = cv2.imread(filename)
    break

# Convert BGR  RGB (for display)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

2. IMAGE HISTOGRAM

```python
plt.figure(figsize=(6,4))
plt.hist(img_rgb.ravel(), bins=256, color='gray')
plt.title("Image Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```

3. CONTRAST STRETCHING

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

min_val = np.min(img_gray)
max_val = np.max(img_gray)

contrast_stretched = ((img_gray - min_val) / (max_val - min_val)) * 255
contrast_stretched = contrast_stretched.astype(np.uint8)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title("Original Gray")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(contrast_stretched, cmap='gray')
plt.title("Contrast Stretched")
plt.axis("off")

plt.show()
```

4. GREY LEVEL CONVERSION

```python
plt.imshow(img_gray, cmap='gray')
plt.title("Gray Level Image")
plt.axis("off")
plt.show()
```

5. IMAGE SMOOTHENING (BLURRING)

```python
smooth = cv2.GaussianBlur(img_gray, (9,9), 0)

plt.imshow(smooth, cmap='gray')
plt.title("Smoothed Image")
plt.axis("off")
plt.show()
```

6. IMAGE SHARPENING

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])

sharpened = cv2.filter2D(img_gray, -1, kernel)

plt.imshow(sharpened, cmap='gray')
plt.title("Sharpened Image")
plt.axis("off")
plt.show()
```

7. NOISE REDUCTION

```python
denoised = cv2.medianBlur(img_gray, 5)

plt.imshow(denoised, cmap='gray')
plt.title("Noise Reduced")
plt.axis("off")
plt.show()
```
